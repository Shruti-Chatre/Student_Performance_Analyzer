```python
student_data["Total_marks"] = student_data["Maths"] + student_data["English"] + student_data["Science"]
```

```python
student_data
```

```python
student_data.sort_values(by = "Total_marks", inplace = True, ascending = False)
```

```python
student_data["Percentage"] = (student_data["Total_marks"]/300)*100
student_data
```

```python
student_data["Percentage"] = round(student_data["Percentage"], 2)
```

```python
student_data
```

```python
student_data["Results"] = student_data["Percentage"].apply(lambda x : "Pass" if x >= 40 else "Fail")
```
